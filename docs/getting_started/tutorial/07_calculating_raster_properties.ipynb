# 07 Calculating Raster Properties


Raster properties such as hillshades for better visualization of the digital elevation model and slope/aspect for the calculation of orientation values can easily be calculated in GemGIS. Hillshades are a 3D grayscale representation of the topography defined by a light source with a given azimuth and altitude. The slope defines the gradient of the raster and the aspect defines the direction of the gradient. Hence, the combination of both paramters results in an orientation value.

## Set File Paths and download Tutorial Data

If you downloaded the latest `GemGIS` version from the Github repository, append the path so that the package can be imported successfully. Otherwise, it is recommended to install `GemGIS` via `pip install gemgis` and import `GemGIS` using `import gemgis as gg`. In addition, the file path to the folder where the data is being stored is set. The tutorial data is downloaded using Pooch (https://www.fatiando.org/pooch/latest/index.html) and stored in the specified folder. Use `pip install pooch` if Pooch is not installed on your system yet.

```python
import gemgis as gg

file_path ='data/07_calculating_raster_properties/'
```

```python
gg.download_gemgis_data.download_tutorial_data(filename="07_calculating_raster_properties.zip", dirpath=file_path)
```

## Loading the raster

```python
import rasterio

raster = rasterio.open(file_path + 'raster.tif')
```

## Plotting the raster

```python
import matplotlib.pyplot as plt

im = plt.imshow(raster.read(1), cmap='gist_earth')
plt.grid()
plt.colorbar(im)
```

## Calculate hillshades

The hillshades can be calculated using ``calculate_hillshades(..)``.

```python
hillshades = gg.raster.calculate_hillshades(raster)
```

## Plotting hillshades

```python
im = plt.imshow(hillshades, cmap='gray')
plt.grid()
plt.colorbar(im)
```

## Calculate slope

The slope can be calculated using ``calculate_slope(..)``.

```python
slope = gg.raster.calculate_slope(raster)
```

## Plotting hillshades

```python
im = plt.imshow(slope, cmap='RdYlGn_r', vmin=0, vmax=90)
plt.grid()
cbar = plt.colorbar(im)
cbar.set_label('Slope in degrees')
```

## Calculate aspect

The aspect can be calculated using ``calculate_aspect(..)``.

```python
aspect = gg.raster.calculate_aspect(raster)
```

## Plotting hillshades

```python
im = plt.imshow(aspect, cmap='twilight_shifted')
plt.grid()
cbar = plt.colorbar(im)
cbar.set_label('Gradient in degrees')
```
